```python
nlp = spacy.load("fr_core_news_md")
```

```python
doc = nlp(corpus)
```

```python
tok = []
prog = ""
n = 6
for token in doc:
    if not token.is_punct and not token.is_stop:
        if len(token.lemma_) > 1 and "\n" not in token.lemma_:
            prog += token.lemma_.lower() + " "
            # tok.append(token.lemma_.lower())
grams = ngrams(prog.split(), n)
for g in grams:
    v1, v2, v3, v4, V5, v6 = g
    txt = v1 + " " + v2 + " " + v3 + " " + v4 + " " + V5 + " " + v6
    tok.append(txt)
```

```python
#order_centroids = model.cluster_centers_.argsort()[:, ::-1]
#terms = X.get_feature_names()
#print(order_centroids[2, 3])
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(tok)
view = dict(zip(vectorizer.get_feature_names(), X.toarray()[0]))
```

```python
# model = KMeans(n_clusters=3)
model = MiniBatchKMeans(n_clusters=3, init='k-means++', max_iter=1000, n_init=10)
model.fit(X)
# plt.scatter(tfidf[:,0],tfidf[:,1], c=model.predict(tfidf))
inertia = []
K_range = range(1,10)
for k in K_range:
    # model = KMeans(n_clusters=k, init='k-means++', n_init=100, max_iter=3000).fit(X)
    model = MiniBatchKMeans(n_clusters=3, init='k-means++', max_iter=1000, n_init=10).fit(X)
    inertia.append(model.inertia_)
plt.plot(K_range, inertia)
```

```python
# Réalisation de la loop
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
MOTS = 1000
for i in range(3):
    print(f"Arbre {str(i+1)}:")
    for a in range(MOTS):
        t = order_centroids[i, a]
        print('- %s' % str(terms[t]), "/", view[terms[t]])
    print("")
```
